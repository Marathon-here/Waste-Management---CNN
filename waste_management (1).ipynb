```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator #for rescale,resize we use ImgaeDatagenerator
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.losses import SparseCategoricalCrossentropy  #required loss function
from tensorflow.keras.optimizers import Adam #required optimizer
```

```python
tensorflow --version
```

```python
train_path = "/content/Waste-Mangement-Dataset/Data/Train"
test_path = "/content/Waste-Mangement-Dataset/Data/Test"
```

#Preprocessing for Train data

```python
train_generator=ImageDataGenerator(rescale=1/255)
train_generator=train_generator.flow_from_directory(
    train_path,
    target_size=(300,300),#This resizes all the images to a fixed size of 300x300 pixels. This is necessary because neural networks require input images of a consistent size.
    batch_size=10,
    class_mode='sparse' #for multiple category classification
)
```

this code gets the mapping of class names to numerical indices from the image data generator and then reverses this mapping to easily interpret the model's predictions.   ----------->only needed for training data

```python
labels = train_generator.class_indices
labels = dict((v,k) for k,v in labels.items())
labels
```

#preprocessing for Test data

```python
test_generator=ImageDataGenerator(rescale=1/255)
test_generator=test_generator.flow_from_directory(
    test_path,
    target_size=(300,300),#This resizes all the images to a fixed size of 300x300 pixels. This is necessary because neural networks require input images of a consistent size.
    batch_size=10,
    class_mode='sparse' #for multiple category classification
)
```

#in two ways we can add layers


```python
model = Sequential()
model.add(Conv2D(32,kernel_size=(3,3),padding="same",activation='relu',input_shape=(300,300,3)))#input layer-conv2d layer(no.of neurons,filters,activation function,input size of 3 chaneels)
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Conv2D(64,kernel_size=(3,3),padding="same",activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Conv2D(128,kernel_size=(3,3),padding="same",activation='relu'))
model.add(Flatten()) #The Flatten layer serves as a bridge between the convolutional/pooling layers and the fully connected layers(Dense layers).
model.add(Dense(32,activation='relu'))
model.add(Dropout(0.2)) #to maintain regularization
model.add(Dense(64,activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(6,activation='softmax')) #output layer- 6 neurons for 6 classes(6-types of categories in our datset)
```

```python
model.compile(optimizer="adam",loss="sparse_categorical_crossentropy",metrics=["accuracy"])
```

```python
model.fit(train_generator,epochs=25,steps_per_epoch=2186//32)
```

```python
model.save("waste_model.h5")
```
